# barplot:
When there are multiple observations in each category, 
it also uses bootstrapping to compute a confidence 
interval around the estimate, which is plotted using error bars.

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
#hue 
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

```python
import numpy as np 
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
```

```python
#ci for removing error bars 
sns.barplot(data=tips, x='sex', y='total_bill',ci=None)
```

```python
#point plot
```

```python
sns.pointplot(data=tips,x='sex', y='total_bill')
```

```python
#hue 
sns.pointplot(data=tips,x='sex', y='total_bill',hue='smoker')
```

```python
#ci = none for removing error bars 
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker', ci=None)
```

```python
#countplot
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

### A second way to plot Facet plots -> FacetGrid

```python
#figure level -> relplot -> displot -> catplot -> lmplot
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
#facet grid 
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
```

#### plotting pairwise realationship (pairgrid vs pairplot)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
#pair grid 
g=sns.PairGrid(data=iris)
# g.map
g.map(sns.scatterplot)
```

```python
#map_diag--> map_offdiag
g=sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
#violinplot
g=sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
#box plot
g=sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
#map_diag-> map_upper()-> map_lower()
```

```python
g=sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### jointGrid Vs Jointplot

```python
sns.jointplot(data=tips, x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist', hue='sex')
```

```python
g=sns.JointGrid(data=tips, x='total_bill',y='tip')
g.plot(sns.scatterplot, sns.histplot)
```

```python
g=sns.JointGrid(data=tips, x='total_bill',y='tip')
g.plot(sns.scatterplot, sns.violinplot)
```

```python
g=sns.JointGrid(data=tips, x='total_bill',y='tip')
g.plot(sns.kdeplot, sns.violinplot)
```
