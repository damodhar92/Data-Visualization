### 2.Distribution Plots

### Plots under distribution plot

```python
# figure level ---> displot
```

```python
# axes level-----> histplot -> kdeplot -> rugplot
```

```python
#plotting univariate histogram
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill',kind='hist')
```

```python
#bins parameter 
sns.displot(data=tips, x='total_bill',kind='hist',bins=2)
```

```python
# It's also possible to visualize the distribution of a categorical variable using logic of histogram.
```

```python
# Discrete bins are automatically set for categorical variables 
```

```python
#count plot this is not histogram conventionally
sns.displot(data=tips,x='day',kind='hist')
```

```python
# hue parameter 
sns.displot(data=tips, x='tip',kind='hist',hue='sex')
```

```python
# element -> step

sns.displot(data=tips, x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
#faceting using column and row -> this function not work on histplot 
sns.displot(data=titanic,x='age',kind='hist',element='step',col='sex')
```

```python
#kde plot 
# Rather than using discrete bins, a KDE plot smooths the observation with a Gassian Kernel,
#->  producing a continuous density estimate.
sns.kdeplot(data=tips,x='total_bill')
```

```python
#hue -> fill 
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
#hue -> fill 
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
#bivariate histogram 

# A bivariate histogram bins the data within rectangles that title the plot 
# and then show the count of observation within each rectangel with the fill color 
```

```python
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
# Bivariate kdeplot
# a bivariate KDE plot smothes the (x,y) observations with a 2D gaussian.
```

```python
sns.kdeplot(data=tips, x='total_bill',y='tip')
```

### Matrix Plot

#### HeatMap
#### ClusterMap

```python
#HeatMap
```

```python
#plot rectangular data as a color encoded matric 
```

```python
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
#axes level function 
sns.heatmap(temp_df)
```

```python
# annot
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True, linewidth=0.5, cmap='viridis')
```

```python
#cluster map

#plot a matrix dataset as a hierarchically-clustered heatmap.
#this function requires scipy 
iris=sns.load_dataset('iris')
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
